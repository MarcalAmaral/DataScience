Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

````python
URL_SP = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
link = requests.get(URL_SP)
````

````python
data_sp = requests.get(URL_SP)
sp_pvl = data_sp.json()
sp_pvl
````

````python
df_pvl_sp = pd.DataFrame(sp_pvl['items'])
df_pvl_sp
````

````python
#1) Seu código aqui
def consulta_pvl (UF: str, tipo_interessado: 'Estado' or 'Municipio' or 'Município' or 'municipio' or 'estado') -> pd.DataFrame:
    
    try:
        if tipo_interessado == 'Estado' or 'estado':
            data = requests.get(f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={UF}&tipo_interessado=Estado')
            data_json = data.json()
            df = pd.DataFrame(data_json['items'])
        else:
            data = requests.get(f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={UF}&tipo_interessado=Munic%C3%ADpio')
            data_json = data.json()
            df = pd.DataFrame(data_json['items'])
            
    except Exception as Exc:
            print(Exc)
        
    return df
````

````python
consulta_pvl(UF = 'SP', tipo_interessado = 'Municipio')
````

````python
# 2) Seu código aqui
df_mg = consulta_pvl(UF = 'MG',
                     tipo_interessado = 'Estado')

print(df_mg['status'].value_counts())
print('\n')
print('A quantidade de solicitações para o Estado de Minas Gerais com o status "Arquivado por decurso de prazo" é de apenas uma.')
````

````python
# 3) Seu código aqui
df_ba = consulta_pvl(UF = 'BA',
                     tipo_interessado = 'Municipio')

df_ba.head()
````

````python
df_ba_deferidos = df_ba[df_ba.status.eq('Deferido')]
df_ba_deferidos
````
