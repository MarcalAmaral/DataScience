```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('./SINASC_RO_2019.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3, 'APGAR5_cat'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5), 'APGAR5_cat'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7), 'APGAR5_cat'] = 'asfixia leve'
df.loc[df['APGAR5'] >= 8, 'APGAR5_cat'] = 'asfixia normal'
```

```python
df['APGAR5_cat'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
labels_gest = list(df['GESTACAO'].unique())
labels_gest
```

```python
labels_gest.pop(5)
```

```python
ax1 = sns.catplot(y = 'APGAR5'
                 ,x = 'GESTACAO'
                 ,hue = 'SEXO'
                 ,data = df
                 ,kind = 'box')

ax1.set_xticklabels(labels_gest, rotation = 60)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
apgar_gest = pd.crosstab(index= df['GESTACAO']
                  ,columns = df['APGAR5']
                  )
apgar_gest
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
apgar_gest_pctlin = apgar_gest.div(apgar_gest.sum(axis=1), axis= 0)
apgar_gest_pctlin
```
